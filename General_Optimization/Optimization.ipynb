```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

import optimizers as op
```

### Ejemplo - $f(x) = x^2$

```python
f = lambda x: x**2

x = np.linspace(-5,5,1000)
plt.plot(x,f(x))
plt.plot(0,0,"*",color="red")
```

#### Búsqueda dicotómica

```python
a,b = op.dichotomic_search(f,(-5,5),0.02,0.05,plot=True)
plt.plot(0,0,"*",color="red",label="True solution")
plt.legend()

print(f"Final solution: [{round(a,3)},{round(b,3)}]")
```

#### Sección aurea

```python
a,b = op.aureus_section(f,(-5,5),0.05,plot=True)
plt.plot(0,0,"*",color="red",label="True solution")
plt.legend()

print(f"Final solution: [{round(a,3)},{round(b,3)}]")
```

## Ejemplo - $f(x,y) = x^2 + y^2$

```python
fp = lambda x,y: x**2+y**2

x = np.linspace(-5,5,1000)
y = np.linspace(-5,5,1000)
X,Y = np.meshgrid(x,y)
Z = fp(X,Y)

plt.contour(X,Y,Z)
plt.plot(0,0,"*",color="red")
```

#### Hooke-Jeeves

```python
f = lambda x: x[0]**2+x[1]**2
x0 = np.array([3,3])

op.hooke_jeeves(f, 0.02, ([-5,-5],[5,5]), x0, plot=True, plot_fobj=fp)
plt.plot(0,0,"*",color="red",label="True solution")
plt.legend()
```

```python
f = lambda x: (x[0]**2 + x[1] - 11)**2 + (x[0]+x[1]**2 -7)**2
fp = lambda x,y: (x**2 + y - 11)**2 + (x+y**2 -7)**2
x0 = np.array([4,10])

print(op.hooke_jeeves(f, 0.02, ([-10,-10],[10,10]), x0, plot=True, plot_fobj=fp))
plt.plot(3,2,"*",color="red",label="True solution")
plt.plot(-2.8,3.1,"*",color="red")
plt.plot(-3.8,-3.3,"*",color="red")
plt.plot(3.6,-1.8,"*",color="red")
plt.legend()
```

#### Gradiente

```python
f = lambda x: x[0]**2+x[1]**2
gradiente_f = lambda x: np.array([2*x[0],2*x[1]])
x0 = np.array([3,3])

op.gradient(f,gradiente_f, 0.02, ([-5,-5],[5,5]), x0, plot=True, plot_fobj=fp)
plt.plot(0,0,"*",color="red",label="True solution")
plt.legend()
```
